### Import the needed Packages

```python
import numpy as np
import matplotlib.pyplot as plt

# Needed for the mnist data
from keras.datasets import mnist
from keras.utils import to_categorical 
```

### Define the class

```python
class NeuralNetwork:

    def __init__(self, ninputs, nlayers, nneurons, noutputs, alpha):
        '''
            layers: List of integers which represents the architecture of the network.
            alpha:  Learning rate.
        '''
        
        # TODO: Initialize the list of weights matrices, then store
        # the network architecture and learning rate
    
             
        self.layers = nlayers
        self.neurons = nneurons
        self.alpha = alpha
        
        self.weighti = np.random.randn(ninputs, nneurons)
        self.weights = np.zeros((nlayers-1, nneurons, nneurons))
        self.weighto = np.random.randn(nneurons,noutputs)
        
        for i in range(nlayers-1):
            self.weights[i] = np.random.randn(nneurons,nneurons)
            
      
        
        self.bi = np.random.randn(nneurons)  
        self.bs = np.random.randn(nlayers-1, nneurons)
        self.bout = np.random.randn(noutputs)
    
            
    def __repr__(self): 
        # construct and return a string that represents the network 
        # architecture 
        return "NeuralNetwork: {}".format( "-".join(str(l) for l in self.layers))

    def softmax(self,X):  
        # applies the softmax function to a set of values
        
        expX = np.exp(X)
        return expX / expX.sum(axis=1, keepdims=True)
        
    def sigmoid(self, x):
        # the sigmoid for a given input value
        
        return 1.0 / (1.0 + np.exp(-x))
    
    def sigmoid_deriv(self, x):
        # the derivative of the sigmoid
        
        return x * (1 - x)
    
    def predict(self, inputs):
        level0 = inputs
        levels = np.zeros((self.layers,inputs.shape[0],self.neurons))
        levels[0] = self.sigmoid(np.dot(level0,self.weighti)+self.bi)
        
        for i in range(1 , self.layers):
            levels[i] = self.sigmoid(np.dot(levels[i-1],self.weights[i-1])+self.bs[i-1])
            
        pred = self.softmax(np.dot(levels[len(levels)-1], self.weighto)+self.bout)

        return pred,levels

    def train(self, inputs, labels, epochs = 1000, displayUpdate = 100):

        # TODO: Define the training step for the network. It should include the forward and back propagation
        # steps, the updating of the weights, and it should print the error every 'displayUpdate' epochs
        # It must return the errors so that they can be displayed with matplotlib

        errors = []

        for epoch in range(epochs):


            pred, levels= self.predict(inputs)

            level0 = inputs
            level1 = levels[len(levels)-1]
            level2 = pred

            level_error = labels-level2


            if epoch%displayUpdate == 0:
                print("Error:", np.mean(np.abs(level_error)))

            errors.append(np.average(np.abs(level_error)))

            level_delta = level_error * self.sigmoid_deriv(pred)

            b_delta = np.sum(level_delta)
            self.bout += b_delta * self.alpha

            level_error = np.dot(level_delta, self.weighto.T)
            self.weighto += np.dot(level1.T,level_delta)*self.alpha 
            level_delta = level_error * self.sigmoid_deriv(level1)

            for i in range(self.layers-1):

                b_delta = np.sum(level_delta)
                self.bs[len(self.bs)-i-1] += b_delta * self.alpha 

                level_error = np.dot(level_delta, self.weights[len(self.weights)-i-1])
                self.weights[len(self.weights)-i-1] += np.dot(levels[len(levels)-i-2].T,level_delta)*self.alpha
                level_delta = level_error * self.sigmoid_deriv(levels[len(levels)-i-2])


            b_delta = np.sum(level_delta)
            self.bi+= b_delta * self.alpha

            self.weighti += np.dot(level0.T,level_delta)*self.alpha
        return errors    
```

### Test datasets

#### XOR

```python
# input dataset
XOR_inputs = np.array([  
                [0,0],
                [0,1],
                [1,0],
                [1,1]
            ])

# labels dataset            
XOR_labels = np.array([[0,1,1,0]]).T
```

```python
#TODO: Test the class with the XOR data

ohlabels = np.zeros((XOR_inputs.shape[0],XOR_inputs.shape[1]))

for i in range(XOR_inputs.shape[0]):
    ohlabels[i,XOR_labels[i]] = 1
print(XOR_inputs.shape[0],XOR_inputs.shape[1])


neuralNetwork = NeuralNetwork(ninputs = 2, nneurons = 4,nlayers = 1, noutputs = 2,alpha = 5)
errors = neuralNetwork.train(XOR_inputs,ohlabels)

#print("dsds",errors)
f, ax = plt.subplots(1,1)
ax.set_xlabel('Epoch')
ax.set_ylabel('Error')
ax.plot(errors)
```

#### Multiple classes

```python
# Creates the data points for each class
class_1 = np.random.randn(700, 2) + np.array([0, -3])  
class_2 = np.random.randn(700, 2) + np.array([3, 3])  
class_3 = np.random.randn(700, 2) + np.array([-3, 3])

feature_set = np.vstack([class_1, class_2, class_3])

labels = np.array([0]*700 + [1]*700 + [2]*700)

one_hot_labels = np.zeros((2100, 3))
print(one_hot_labels[0])
for i in range(2100):  
    one_hot_labels[i, labels[i]] = 1

plt.figure(figsize=(10,10))  
plt.scatter(feature_set[:,0], feature_set[:,1], c=labels, s=30, alpha=0.5)  
plt.show()
```

```python
#TODO: Test the class with the multiple classes data

neuralNetwork = NeuralNetwork(ninputs = 2, nneurons = 3,nlayers = 1, noutputs = 3,alpha = .01)

errors = neuralNetwork.train(feature_set,one_hot_labels)


f, ax = plt.subplots(1,1)
ax.set_xlabel('Epoch')
ax.set_ylabel('Error')
ax.plot(errors)
```

#### On the mnist data set

---
Train the network to classify hand drawn digits.

For this data set, if the training step is taking too long, you can try to adjust the architecture of the network to have fewer layers, or you could try to train it with fewer input. The data has already been loaded and preprocesed so that it can be used with the network.

---

```python
# Load the train and test data from the mnist data set
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Plot a sample data point
plt.title("Label: " + str(train_labels[0]))
plt.imshow(train_images[0], cmap="gray")
```

```python
# Standardize the data

# Flatten the images
train_images = train_images.reshape((60000, 28 * 28))
# turn values from 0-255 to 0-1
train_images = train_images.astype('float32') / 255 

test_images = test_images.reshape((10000, 28 * 28)) 
test_images = test_images.astype('float32') / 255

# Create one hot encoding for the labels
train_labels = to_categorical(train_labels) 
test_labels = to_categorical(test_labels)
```

```python
# TODO: Test the class with the mnist data. Test the training of the network with the test_images data, and 
# record the accuracy of the classification.
neuralNetwork =  NeuralNetwork(ninputs = 784, nneurons = 64,nlayers = 2, noutputs =10,alpha = 0.0005)


errors = neuralNetwork.train(train_images[0:5000],train_labels[0:5000])


f, ax = plt.subplots(1,1)
ax.set_xlabel('Epoch')
ax.set_ylabel('Error')
ax.plot(errors)
```

```python
tests = neuralNetwork.predict(test_images[0:1000])[0]
one_hot_labels = to_categorical(test_labels[0:1000])

predictions = np.copy(tests)
predictions[predictions > 0.5] = 1
predictions[predictions < 0.5] = 0

errorImages = []
for index, (prediction, label) in enumerate(zip(predictions[0:10], one_hot_labels[0:10])):
    if not np.array_equal(prediction,label):
        errorImages.append((index, prediction, label))

f, plots = plt.subplots(10, 1, figsize=(50,50))

for img, plot,lab in zip(errorImages, plots,one_hot_labels):
    plot.imshow(test_images[img[0]].reshape(28,28), cmap = "gray")
    plot.set_title('\n \n    Label:' + str(lab.T[1]) + '\nPrediction: ' + str(img[1]) )
```
